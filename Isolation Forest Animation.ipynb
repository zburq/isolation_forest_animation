# Isolation Forest Animation

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import random
from intervals import openclosed as oc
```

```python
def generate_split():
    d = random.choice(['x', 'y'])
    split_loc = random.uniform(0,10)
    return [d, split_loc]

def list_to_intervals(lst):
    lst = sorted(lst + [0,10] ) 
    return [ oc(a,b) for a,b in zip(lst[:-1], lst[1:]) ]

def splits_to_intervals(splits):
    xx = list_to_intervals([ s[1] for s in splits if s[0] == 'x'])
    yy = list_to_intervals([ s[1] for s in splits if s[0] == 'y'])
    return {'x_intervals': xx, 'y_intervals': yy}

def splits_to_lines(splits):
    x_lines = [ plt.axline( (p[1], 0), (p[1], 1)) for p in splits if p[0] == 'x']
    y_lines = [ plt.axline( (0, p[1]), (1, p[1])) for p in splits if p[0] == 'y']
    return {'x_lines': x_lines, 'y_lines': y_lines}

def check_isolation(point, data, splits):
    intervals = splits_to_intervals(splits)
    x_int = [i for i in intervals['x_intervals'] if point[0] in i][0]
    y_int = [i for i in intervals['y_intervals'] if point[1] in i][0]
    caught_points = [ p for p in data.values if p[0] in x_int and p[1] in y_int ]    
    return len(caught_points)
```

```python
point_1 = [3,6]
point_2 = [5,5]
x = [round(random.uniform(0,10),2) for _ in range(25)]
y = [ a + round(random.uniform(-1,1),2) for a in x]
data = pd.DataFrame([x,y]).transpose().rename(columns = {0:'x', 1:'y'})
data.loc[25] = point_1
data.loc[26] = point_2
data = data[(data < 10)&(data > 0)].dropna()
data
```

```python
point = point_2
split_counts = []
m = 0
for j in range(5):
    cell_content = 3
    splits = [['x',0],['x',10],['y',0],['y',10]]
    i = 0
    if m >  0:
        n = m
    else: n = ''
    while cell_content > 1:
        plt.figure(figsize=(5,5))
        data.plot.scatter(x  = 'x', y = 'y', figsize = (5,5))
        cell_content = check_isolation(point, data, splits)
        if cell_content > 1:
            [s, ec, fc] = [100 ,'r', 'none']
        else:
            [s, ec, fc] = [300,'k','r']
        plt.scatter( point[0],point[1], s=s, facecolors=fc, edgecolors=ec)
        plt.xlim(-.10,10.1)
        plt.ylim(-.1,10.1)
        splits_to_lines(splits[4:])
        plt.title(f'Loop:{j+1}\nSplit Count: {i}\nSplit Count Average: {n}')
        plt.tight_layout()
        plt.savefig(f'./output/isolation_forest_gif_{j+1}{i+1:03}.jpg', dpi = 100)   
        splits.append(generate_split())
        i = i+1        
    split_counts.append(i-1)
    m = round(np.mean(split_counts),2)

plt.figure(figsize=(5,5))
data.plot.scatter(x  = 'x', y = 'y', figsize = (5,5))
plt.scatter( point[0], point[1], s=100, facecolors='none', edgecolors='r')
plt.xlim(-.10,10.1)
plt.ylim(-.1,10.1)
plt.title(f'Loops run:{j+1}\nSplit Count Average: {m}')
plt.tight_layout()
plt.savefig(f'./output/isolation_forest_gif_{j+2}000.jpg',  dpi = 100)
    #    print(cell_content) 
```
